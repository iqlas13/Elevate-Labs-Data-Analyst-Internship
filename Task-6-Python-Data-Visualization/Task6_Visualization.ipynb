## Importing Required Libraries

In this step, the required Python libraries for data visualization are imported.
Pandas is used for data handling and Matplotlib is used to create different types
of charts for visual analysis.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Loading the Dataset

In this step, the cleaned Medical Insurance dataset is loaded into a Pandas
DataFrame. Using the cleaned dataset ensures accurate and reliable visualizations.


```python
df = pd.read_csv("cleaned_insurance.csv")
```

```python
df.head()
```

## Bar Chart: Average Insurance Charges by Smoking Status

A bar chart is used to compare values across different categories. In this
visualization, we compare the average insurance charges for smokers and
non-smokers to understand how smoking impacts insurance cost.


```python
avg_charges_smoker = df.groupby('smoker')['charges'].mean()

plt.figure(figsize=(6,4))
avg_charges_smoker.plot(
    kind='bar',
    color=['skyblue', 'salmon']   # colors added
)

plt.title('Average Insurance Charges by Smoking Status')
plt.xlabel('Smoking Status')
plt.ylabel('Average Charges')
plt.grid(axis='y')

# save the figure
plt.savefig("bar_chart_smoker_vs_charges.png", bbox_inches='tight')

plt.show()
```

## Line Chart: Insurance Charges vs Age

A line chart is useful for observing trends and patterns over a continuous
variable. In this visualization, we analyze how insurance charges change with
respect to age.


```python
# Sort data by age for a proper line plot
df_sorted = df.sort_values(by='age')

plt.figure(figsize=(7,4))
plt.plot(df_sorted['age'], df_sorted['charges'], color='green')
plt.title('Insurance Charges vs Age')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.grid(True)

# Save the plot
plt.savefig("line_chart_age_vs_charges.png", bbox_inches='tight')

plt.show()
```

## Histogram: Distribution of Insurance Charges

A histogram is used to understand the distribution of a numerical variable.
In this visualization, we analyze how insurance charges are distributed across
different ranges.


```python
plt.figure(figsize=(7,4))
plt.hist(df['charges'], bins=30, color='purple', edgecolor='black')
plt.title('Distribution of Insurance Charges')
plt.xlabel('Charges')
plt.ylabel('Frequency')
plt.grid(True)

# Save the plot
plt.savefig("histogram_insurance_charges.png", bbox_inches='tight')

plt.show()
```

## Scatter Plot: BMI vs Insurance Charges

A scatter plot is used to study the relationship between two numerical variables.
In this visualization, we examine the relationship between BMI and insurance
charges to understand whether higher BMI leads to higher medical insurance costs.


```python
plt.figure(figsize=(7,4))
plt.scatter(df['bmi'], df['charges'], color='orange', alpha=0.6)
plt.title('BMI vs Insurance Charges')
plt.xlabel('BMI')
plt.ylabel('Charges')
plt.grid(True)

# Save the plot
plt.savefig("scatter_bmi_vs_charges.png", bbox_inches='tight')

plt.show()
```
